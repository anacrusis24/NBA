## 2 step process
 1. add W/L and remove unwanted columns from each game df
 2. stack each cleaned game df into the pca_data 

```python
box = boxscoretraditionalv2.BoxScoreTraditionalV2(game_id = '0022101220')
df = box.get_data_frames()[0]
```

```python
season_2017 = season_2017[['GAME_ID', 'TEAM_ID', 'TEAM_NAME', 'WL']]
```

```python
box = boxscoretraditionalv2.BoxScoreTraditionalV2(game_id = season_2017.GAME_ID.unique())
```

```python
pca_data = pd.DataFrame()


for game_id in tqdm(season_2017.GAME_ID.unique(), 'GAMES'):
    box = boxscoretraditionalv2.BoxScoreTraditionalV2(game_id = game_id)
    df = box.get_data_frames()[0]
    pca_data = pd.concat([pca_data, df])
    time.sleep(1)

pca_data.to_csv('pca_data_unclean.csv')  
```

```python
pca_data.columns
```

```python
pca_data = pca_data[['GAME_ID', 'TEAM_ID', 'MIN', 'FG_PCT', 'FGA', 'FG3_PCT', 'FG3A', 'FT_PCT', 'FTA',
                     'OREB', 'DREB', 'AST', 'STL', 'BLK', 'TO', 'PTS', 'PF']]
```

```python
pca_data.set_index(['GAME_ID', 'TEAM_ID'], inplace=True)
season_2017.set_index(['GAME_ID', 'TEAM_ID'], inplace=True)

final_df = pca_data.join(season_2017).reset_index()
```
